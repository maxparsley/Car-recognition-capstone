```python
import os
import shutil
# Renaming, copying, and removing the first 4816 files to the training folder
source_folder = 'G:\cars test resized - Copy'
destination_folder = 'G:\cars train resized - Copy'
start_number = 8145

# Create the destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# Get the list of files in the source folder
files = os.listdir(source_folder)

# Sort the files to ensure consistent ordering
files.sort()

# Iterate through the first 4816 files
for i, file_name in enumerate(files[:4816]):
    # Generate the new file name with sequential numbers
    new_file_name = f'{start_number + i:05d}.jpg'

    # Build the full paths for the source and destination
    source_path = os.path.join(source_folder, file_name)
    destination_path = os.path.join(destination_folder, new_file_name)

    # Rename and copy the file to the destination folder
    shutil.copyfile(source_path, destination_path)

    # Remove the original file from the source folder
    os.remove(source_path)

print('Renaming, copying, and removing completed.')
```
